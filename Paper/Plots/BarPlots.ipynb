```python
import numpy as np
import matplotlib as mpl
import matplotlib.patches as ptc
mpl.rcParams['figure.dpi'] = 1000
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(5,3))
N = 4
ind = np.arange(N)  # the x locations for the groups
width = 0.4      # the width of the bars
plt.ylim([0, 0.25])

fig = plt.figure()

yvals = [0.15163175,0.158774604,0.153988636,0.168439529]
rects1 = ax.bar(ind, yvals, width, color='#da8493')
zvals = [0.133089077,0.139888239,0.140881799,0.146821631]
rects2 = ax.bar(ind+width, zvals, width, color='#ef8b6d')

ax.set_ylabel('Error Value (RMSD)',fontsize=15)
ax.set_xlabel('Bend Numbers',fontsize=15)

ax.set_xticks(ind+width)
ax.set_xticklabels( ('1', '2', '3', '4') )
ax.legend( (rects1[0], rects2[0]), ('Deterministic Error', 'Probabilistic Error'),frameon=False,prop={'size': 15})

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(5,3))
N = 6
ind = np.arange(N)  # the x locations for the groups
width = 0.4      # the width of the bars

fig = plt.figure()

yvals = [5558/6400, 5946/6400, 6867/6400,10115/6400,9603/6400,10207/6400]
rects1 = ax.bar(ind, yvals, width, color='#3e50c4')
zvals = [2995/6400,5656/6400,9721/6400,10059/6400,15902/6400,19964/6400]
rects2 = ax.bar(ind+width, zvals, width, color='#aec8fd')

ax.set_ylabel('Percentage Area',fontsize=15)
ax.set_xlabel('Year',fontsize=15)

ax.set_xticks(ind+width)
ax.set_xticklabels( ('5', '10', '15', '20', '25', '30') )
ax.legend( (rects1[0], rects2[0]), ('Uncertainty (0.25<x<0.75)', 'Erosion (0.75<x<1.0)'),frameon=False,prop={'size': 15})

y_value=['{:,.1f}'.format(x) + '%' for x in ax.get_yticks()]
ax.set_yticklabels(y_value)
plt.show()
```
