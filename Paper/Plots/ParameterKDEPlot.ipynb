```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
import meanderpy as mp                   # https://github.com/zsylvester/meanderpy
from importlib import reload
reload(mp)
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as ptc
mpl.rcParams['figure.dpi'] = 1000
import scipy.stats as stats             # for distributions
import seaborn as sns                   # for 3D plotting
from scipy import interpolate           # for interpolation
import random
os.chdir(r"/Users/braydennoh/Downloads/synth")
samp2d1=np.loadtxt('synethbend1.txt')
samp2d2=np.loadtxt('synethbend2.txt')
samp2d3=np.loadtxt('synethbend3.txt')
samp2d4=np.loadtxt('synethbend4.txt')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks")
fig, ax = plt.subplots(figsize=(4,4))
d1= samp2d1[:, 0]
d2= samp2d2[:, 0]
d3= samp2d3[:, 0]
d4= samp2d4[:, 0]

# plotting both distibutions on the same figure
fig = sns.kdeplot(d1,  color="#da8493", alpha=.8, linewidth=0,fill=True,label='1 bend')
fig = sns.kdeplot(d2, color="#ef8b6d", alpha=.7, linewidth=0,fill=True,label='2 bends')
fig = sns.kdeplot(d3, color="#aec8fd", alpha=.6, linewidth=0,fill=True,label='3 bends')
fig = sns.kdeplot(d4, color="#3e50c4", alpha=.5, linewidth=0,fill=True,label='4 bends')
plt.axvline(x=20, color='black', label='Truth Value',ls='--', lw=3)

plt.xlim([0, 40])
plt.ylim([0, 0.2])

fig.legend(frameon=False,prop={'size': 11})
plt.xlabel('Migration Rate',fontsize=15)
plt.ylabel('Density',fontsize=15)
plt.show()

```
